# Forecasting Car Price using Multiple Linear Regression

**Vấn đề:**

Giả sử VinFast muốn thâm nhập thị trường Mỹ bằng cách thành lập đơn vị sản xuất của họ ở đó và sản xuất ô tô tại địa phương để cạnh tranh với các đối tác Mỹ và châu Âu.

Họ đã ký hợp đồng với một công ty tư vấn ô tô để tìm hiểu các yếu tố mà việc định giá ô tô phụ thuộc vào đó. Cụ thể, họ muốn hiểu các yếu tố ảnh hưởng đến việc định giá ô tô tại thị trường Mỹ, vì những yếu tố này có thể rất khác so với thị trường nơi họ hiện đang hoạt động. Công ty đó muốn biết:
+ Biến số nào có ý nghĩa trong việc dự đoán giá xe ô tô
+ Những biến đó mô tả giá của một chiếc ô tô tốt như thế nào

Dựa trên các cuộc khảo sát thị trường khác nhau, công ty tư vấn đã thu thập được một bộ dữ liệu lớn về các loại ô tô khác nhau trên thị trường châu Mỹ.

**Mục tiêu:**

Lập mô hình về giá ô tô với các biến độc lập có sẵn. Nó sẽ được ban quản lý sử dụng để hiểu chính xác giá thay đổi như thế nào với các biến độc lập. Theo đó, họ có thể điều chỉnh thiết kế của ô tô, lên chiến lược kinh doanh, ...để đáp ứng các mức giá nhất định. Hơn nữa, mô hình này sẽ là một cách tốt để ban quản lý hiểu được động lực định giá của một thị trường mới.

Dưới đây là quy trình phân tích đề xuất.

## Step 4 : Deriving new features (Phát sinh các tính năng mới)

```python
# Fuel economy (Tiết kiệm nhiên liệu)
cars['fueleconomy'] = (0.55 * cars['citympg']) + (0.45 * cars['highwaympg'])
```

```python
# Binning các công ty ô tô dựa trên giá trung bình của mỗi công ty.
cars['price'] = cars['price'].astype('int')
temp = cars.copy()
table = temp.groupby(['CompanyName'])['price'].mean()
temp = temp.merge(table.reset_index(), how='left',on='CompanyName')
bins = [0,10000,20000,40000]
cars_bin=['Budget','Medium','Highend']
cars['carsrange'] = pd.cut(temp['price_y'],bins,right=False,labels=cars_bin)
cars.head()
```

## Step 6 : Dummy Variables (Biến giả)

```python
# Defining the map function
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df
# Áp dụng function cho cars_lr

cars_lr = dummies('fueltype',cars_lr)
cars_lr = dummies('aspiration',cars_lr)
cars_lr = dummies('carbody',cars_lr)
cars_lr = dummies('drivewheel',cars_lr)
cars_lr = dummies('enginetype',cars_lr)
cars_lr = dummies('cylindernumber',cars_lr)
cars_lr = dummies('carsrange',cars_lr)
```

## Step 7 : Model Building

### Cách 1:

```python
cars_lr.columns
```

```python
X = cars_lr[['wheelbase', 'curbweight', 'enginesize', 'boreratio',
       'horsepower', 'fueleconomy', 'carlength', 'carwidth', 'gas', 'turbo',
       'hardtop', 'hatchback', 'sedan', 'wagon', 'fwd', 'rwd', 'dohcv', 'l',
       'ohc', 'ohcf', 'ohcv', 'rotor', 'five', 'four', 'six', 'three',
       'twelve', 'two', 'Medium', 'Highend'
       ]]
y = cars_lr['price']
```

### Train-Test Split and feature scaling

```python
# Chuẩn bị dữ liệu training data/testing data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.3, random_state = 42)
```

```python
# Load module, tạo đối tượng linear regression
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression ()
model.fit (X_train, y_train)

# Để dự báo, dùng model.predict
y_pred = model.predict (X_test)
```

```python
intercept = model.intercept_
slope = model.coef_
print (intercept, slope)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
def Static_score_model_reg (y, y_pred):
    r2 = r2_score (y, y_pred)
    mse = mean_squared_error (y, y_pred)
    mae = mean_absolute_error (y, y_pred)
    return r2, mse, mae
```

```python
def Visualize_model_reg (y, y_pred):
    plt.figure (figsize=(12, 6))
# Scatter plot    
    plt.subplot (1, 2, 1)
    plt.scatter (y_pred, y)
    plt.xlabel ("Model Predictions")
    plt.ylabel ("True Value")
# Đặt giới hạn trục    
    plt.plot ([0, np.max(y) + 2*np.min(y)], [0, np.max(y) + 2*np.min(y)], '-', color = "r")
#  Displot   
    plt.subplot (1, 2, 2)
    sns.distplot (y, hist = False, color = "r", label = "True Value")
    sns.distplot (y_pred, hist = False, color = "b", label = "Model Predictions", axlabel="Distribution")
    plt.legend()
    plt.show()
```

### Xét trên tập Train

```python
r2, mse, mae = Static_score_model_reg (y_train, model.predict (X_train))
print ('Train score: ', r2)
print ('Train MSE: ', mse)
print ('Train MAE: ', mae)
```

```python
Visualize_model_reg (y_train, model.predict (X_train))
```

### Xét trên tập test

```python
r2, mse, mae = Static_score_model_reg (y_test, y_pred)
print ('Test score: ', r2)
print ('Test MSE: ', mse)
print ('Test MAE: ', mae)
```

```python
Visualize_model_reg (y_test, model.predict (X_test))
```

**Nhận xét:**
+ Dựa trên kết quả ở tập train và tập test, ta thấy model bị **overfitting**.
+ Do đó, model vẫn chưa tối ưu, cần tìm cách khắc phục.

**Cách khắc phục đề xuất đó là lựa chọn lại feature phù hợp với model.**

Tiếp theo, ta sẽ sử dụng chức năng `LinearRegression` từ SciKit Learn để tương thích với `RFE` (là một tiện ích từ `sklearn`).

Approach:
+ Trước tiên, chúng ta sẽ sử dụng `RFE` (Recursive feature elimination) để xác định `top 10` features ảnh hưởng nhiều nhất đến giá xe.
+ Khi điều này được thực hiện, chúng tôi sẽ sử dụng phương pháp `Mixed` trong đó chúng tôi sẽ kiểm tra cả `p-value` và `VIF` để xác định biến nào cần giữ lại và biến nào cần loại bỏ khỏi mô hình.

### RFE
Recursive feature elimination

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train,y_train)
rfe = RFE(lm, n_features_to_select=10) # Số lượng biến chọn = 10 (n_features_to_select=10)
rfe = rfe.fit(X_train, y_train)
```

```python
# RFE function will now determine the ranking of all the variables and rank them for our use.
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# Here we are taking the top 10 columns which are recommended by the RFE function
col = X_train.columns[rfe.support_]
col
```

```python
# In các cột đang bị loại bỏ khỏi phân tích sâu hơn
X_train.columns[~rfe.support_]
```

### Mixed Approach:
Bây giờ chúng ta đã có 10 biến số hàng đầu có ảnh hưởng lớn nhất đến giá ô tô, bây giờ chúng ta sẽ tiếp tục với cách tiếp cận hỗn hợp.

```python
# Tạo X_test dataframe với các biến được chọn RFE
X_train_rfe = X_train[col]
```

**Lựa chọn và loại biến khỏi Model dựa trên:**
+ Điều kiện 1: `p=value` (<0.05)
+ Điều kiện 2: VIF < 5

**Quy trình gồm:** 

Để loại bỏ từng biến một, ta dựa trên các tiêu chí dưới đây:

+ Giá trị `p-value` cao, `VIF` cao : **Bỏ biến**
+ Giá trị `p-value` cao, `VIF` thấp hoặc `p-value` thấp `VIF` cao : **Loại bỏ biến có `p-value` cao trước**
+ Giá trị `p-value` thấp, `VIF` thấp : **chấp nhận biến**

### Building model using statsmodel, for the detailed statistics

```python
import statsmodels.api as sm
```

```python
X_train_rfe = sm.add_constant(X_train_rfe)
X_train_rfe.head()
```

```python
def build_model(X,y):
    X = sm.add_constant(X) #Adding the constant
    lm = sm.OLS(y,X).fit() # fitting the model
    print(lm.summary()) # model summary
    return X
```

```python
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

**Nhận xét:**
+ Các giá trị p-value phù hợp.
+ Tiếp theo, kiểm tra VIF.

**Kiểm tra VIF:**

Variance Inflation Factor (VIF), đưa ra ý tưởng định lượng cơ bản về mức độ tương quan của các biến đặc trưng với nhau. Nó là một tham số cực kỳ quan trọng để kiểm định mô hình tuyến tính. Công thức tính `VIF` là:

$$ \text{VIF} = \frac{1}{1-R^2} $$

```python
checkVIF(X_train_new)
```

**Nhận xét:** Chỉ số VIF của các biến lớn hơn 5 cần loại bỏ biến khỏi model gồm:
